### Visualisation du champ de vecteurs

Pour visualiser un champ de vecteurs, on utilise les commandes suivantes:
- `meshgrid` qui à partir de données $x$, $y$ (array de valeurs) créé une grille de maillage $(X, Y)$ pour le tracé du champ
- `quiver` qui sert à tracer une flèche, de composantes $F_x$ et $F_y$ au point de coordonnées de la grille $(X, Y)$.

Remplacer les `?` du code ci-dessous pour tracer le champ de vecteur $\vec{F}(x, y)$. Commentaires ?

```python
# Création du maillage
x    = np.linspace(-5, 5, 10)
y    = np.linspace(-5, 5, 10)
X, Y = np.meshgrid(x, y) 

# u et v, composantes du champ F respectivement suivant les coordonnées x et y
u = -x
v = y

# Tracé du champ de vecteurs
figure, ax = plt.subplots(1,1,figsize=(6,6)) 
ax.quiver(u, v, X, Y, color='r')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Champ de vecteurs')
ax.grid()
```

## 2. Première application

Utiliser les étapes vues précédemment pour représenter le champ $\vec{F}(x,y,z) = (2 - 2x) \vec{i} + 4 y \vec{j}$, son potentiel (avec $V(0)=0$), ses équipotentielles et ses lignes de champ.

*Réponse*
- Equation donnnée par $4ydx-2(1-x)dy = 0$ =>  $y = \frac{K}{(1-x)^2}$, où $K$ est une constante
- Potentiel $-grad(V) = \vec{F}$ => $V(x,y)  = -x(2-x) - 2y^2$

```python
# Création du maillage
x    = np.linspace(-10, 10, 30)
y    = np.linspace(-10, 10, 30)
X, Y = np.meshgrid(x, y) 

# u et v, composantes du champ F respectivement suivant les coordonnées x et y
u = 2-2*x
v = 4*y

# Tracé du champ de vecteurs
figure, ax = plt.subplots(1,1,figsize=(6,6)) 
ax.quiver(u, v, X, Y, color='r')
ax.set_title('Champ de vecteurs')

x    = np.linspace(-10, 10, 100)
y    = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(x, y) 

# Création du maillage
a = -10
b = 10

# Définition du potentiel
Vp = -X*(2-X)-2*Y**2

# Tracé du Potentiel
c  = 30
ax.contour(X,Y, Vp, c)
ax.set_xlabel('X')
ax.set_ylabel('Y')

#Tracé des lignes de champ
# K constante
for K in range(-20,20,5):  
    ax.plot(x, K/(1-x)**2, color='r')  
ax.set_xlim([a,b])
ax.set_ylim([a,b])
ax.grid()
```

## 3. Champ de vecteurs et équations différentielles

### Définiton

Soit l'équation différentielle $y'(t)=f(t,y(t))$. On note le point du plan $M(t,y)$ d'abscisse $t$ et d'ordonnée $y$. Si la valeur $f(t,y)$ est bien définie, on peut définir un champ de vecteur associé à cette équation différentielle, encore appelé le **champ des pentes** : 

$$\vec T(t,y)=1\vec i+f(t,y)\vec j$$

Par définition, toute courbe d'une solution de l'équation est alors tangente en chacun de ses points au vecteur correspondant. En pratique on quadrille une portion du plan et en chaque point $M$ de ce quadrillage, on trace le vecteur $\vec T$ de norme donnée, pour que le graphe soit lisible. **Les courbes des solutions de l'équation différentielle sont alors les lignes de champ.**
- Intérêt : Si on ne sait pas résoudre l'équation différentielle, la carte de champ de $\vec T$ donne l'allure des solutions.
- Intérêt (bis) : cela prolonge la méthode d'Euler, puisque en chaque point on trace le vecteur tangent qui nous avais permis, au TP précédent, de tracer une solution approchée par la méthode d'Euler. Ici, c'est comme si on obtenait le graphe de toutes les courbes solutions d'un coup !




### Application pour l'équation  $y' = t - 2y$

- Montrer que $f(t)=\frac 12 t-\frac 14$ est solution particulière de l'équation différentielle $y'(t)=t-2y(t)$, puis donner toutes les solutions. 
- Représenter le champ des pentes ainsi que quelques solutions sur le même graphe. 


```python
def f(t,y):
    return t-2*y

def sol(t,k):
    # Solution de l'équation différentielle
    return k*np.exp((-2*t))+t/2-1/4

t    = np.linspace(-10, 10, 30)
y    = np.linspace(-10, 10, 30)
T, Y = np.meshgrid(t, y) 

u = 1
v = f(T,Y)

t1 = np.linspace(-10, 10, 100)

# Tracé du champ de vecteurs
figure, ax = plt.subplots(1,1,figsize=(6,6)) 
ax.quiver(T, Y, u, v, color='r')
ax.set_title('Champ de vecteurs')

for k in range(10):
    ax.plot(t1, sol(t1,k), label='k='+str(k))
ax.legend()
ax.set_xlim([-5,5])
ax.set_ylim([-5,5])
```
