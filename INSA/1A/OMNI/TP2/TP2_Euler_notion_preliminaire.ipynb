## Equation différentielle

Soit l'équation différentielle non linéaire suivante :

$$
\dot u(t) =  f(u(t))
 $$

où $f$ et $u$ sont des fonctions de $\mathbb R$ dans $\mathbb R$.

## Stabilité, instabilité 

La fonction $u$ atteint un **point d'équilibre** $u_e$ lorsque sa dérivée s'annule, c.a.d. quand $\dot u(t) =0$ soit $f(u(t))=0$.

Un point d'équilibre est soit **attractif (stable)** soit **répulsif (instable)**, cf. figure ci-dessous. Si au voisinage de $u_e$,
- $\dot u(t) = f(u(t))$ est négatif pour $u<u_e$ et positif pour $u>u_e$ alors l'équilibre est répulsif ou instable, la population $u$ aura tendance à s'éloigner de la valeur $u_e$. 
- $\dot u(t) = f(u(t))$ est positif pour $u<u_e$ et négatif pour $u>u_e$ alors l'équilibre est stable, la population $u$ aura tendance à s'approcher de la valeur $u_e$.
- $\dot u(t) = f(u(t))$ s'annule sans changer de signe, le point d'équilibre est ni stable ni instable.

## Potentiel

On peut introduire également la notion de **potentiel** $V:\mathbb R \to \mathbb R$ associé à l'équation différentielle : 
    
$$V'(u(t)) = -f(u(t)).$$ 
    
Les points d'équilibre sont donc les points **critiques** de $V$, c'est-à-dire les points où $V'$ s'annule. Un point d'équilibre stable correspondra à un minimum local de $V$, alors qu'un équilbre instable correspondra à un maximum local de $V$.

## Exemples simples

On considère les deux équations différentielles suivantes :

$$\dot u(t) = f_1(u) = u-u^2$$

$$\dot u(t) = f_2(u) = u^2-u^3$$

On en détermine respectivement les deux potentiels $V_1$ et $V_2$ associés :

$$V_1(u)=-\frac{u^2}2 + \frac{u^3}3+C_1$$

$$V_2(u)=-\frac{u^3}3 + \frac{u^4}4+C_2$$

On peut alors tracer dans un même repère la fonction et son potentiel associé. On choisit les constantes d'intégrations $C_1$ et $C_2$ et les échelles pour que les graphes soient lisibles. 

```python
# Import
import matplotlib.pyplot as plt
import numpy as np

# Fonctions
def f1(u):
    return u - u**2

def f2(u):
    return u**2 - u**3

def V1(u, c):
    return -u**2/2 + u**3/3 + c

def V2(u, c):
    return -u**3/3 + u**4/4 + c

# On utilise ici la bibliothèque Numpy pour créer un vecteur. 
# np.linspace(start, stop, num) génère un tableau de nombres équidistants entre start et stop, avec num éléments.
# Ici, start = -2, stop = 2 et num = 100.
u = np.linspace(-2, 2, 100)  

# Graphes
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 4))
ax1.grid(True)
ax1.plot(u, f1(u), label = 'fonction $f_1(u)$')
ax1.plot(u, V1(u,1), label = 'fonction $V_1(u)$') # c=1  pour éviter le chevauchement
ax1.set_xlabel('u')
ax1.set_ylabel('$f_1(u)$ et $V_1(u)$')
ax1.set_title('Cas 1')
ax1.axis([-2,2,-3,2])
ax1.legend()
ax2.grid(True)
ax2.plot(u, f2(u), label = 'fonction $f_2(u)$')
ax2.plot(u, V2(u,0.5), label = 'fonction $V_2(u)$') 
ax2.set_xlabel('u')
ax2.set_ylabel('$f_2(u)$ et $V_2(u)$')
ax2.set_title('Cas 2')
ax2.axis([-2,2,-0.5,1])
ax2.legend()
```

- Dans le cas 1, il y a deux points d'équilibre en $u = 0$ et $u = 1$. $u =0 $ est un maximum local de $V_1$ donc instable, $u = 1$ est un minimum local de $V_1$ donc stable. 
- Dans le cas 2, il y a deux points d'équilibre en $u = 0$ et $u = 1$. $f_2$ ne change pas de signe au voisinage de $u = 0$, donc $V_2$ n'a pas d'extremum, donc le point déquilibre est ni stable ni instable. Pour $u = 1$ on a un minimum local de $V_2$ donc stable.
