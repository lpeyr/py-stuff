### Exercice 2 - Affichage de courbes

<u>Rappel :</u> Un guide de démarrage Matplotlib est disponible sur la page ISN - Ressources Communes sur Moodle : [version pur Python](https://moodle.insa-lyon.fr/pluginfile.php/421597/mod_resource/content/0/Matplotlib_pur_Python.pdf)  ou [version Numpy](https://moodle.insa-lyon.fr/pluginfile.php/421598/mod_resource/content/0/Matplolib_Python_Numpy.pdf) 

Le code ci-dessous, issu du TD2, utilise des listes python pour tracer une portion de la fonction $f(x) = x^2$.  Toutes les abscisses sont regroupées dans une liste (`x`) et les ordonnées correspondantes le sont dans une AUTRE liste (`fx`).

```python
# Affichage de la courbe $y = x^2$ 


x = [0, 1, 2, 3, 4, 5] # Liste des valeurs sur l'axe des abscisses
fx = [0, 1, 4, 9, 16, 25] # Liste des valeurs sur l'axe des ordonnées

fig, ax = plt.subplots() # Création de la figure (fig) et des axes (ax) de la figure
ax.plot(x, fx)           # Création de la courbe sur les axes de la figure

ax.set(xlabel='x', ylabel='f(x)', title='$y = x^2$ ')  # Ajout de la légende des axes et d'un titre
plt.show()  # Affichage de la figure 
```

---
On souhaite afficher la fonction $y=x^2$ de manière plus précise sur l'intervalle $[-10, 10]$.
Pour cela on propose d'utiliser la fonction `linspace(start, stop, nb_points)` de la bibliothèque `numpy`. Cette fonction renvoie un nd-array contenant `nb_points` réparties uniformément entre `start` et `stop`. Documentation de la fonction `linspace`: https://numpy.org/doc/stable/reference/generated/numpy.linspace.html


#### Q4. Utilisez la fonction `linspace` afin d'afficher la courbe de la fonction $y=x^2$ sur l'intervalle $[-10, 10]$ sur $100$ points.

```python
x = np.linspace(-10, 10, 100) # Liste des valeurs sur l'axe des abscisses
fx = x**2 # Liste des valeurs sur l'axe des ordonnées

fig, ax = plt.subplots() # Création de la figure (fig) et des axes (ax) de la figure
ax.plot(x, fx)           # Création de la courbe sur les axes de la figure

ax.set(xlabel='x', ylabel='f(x)', title='$y = x^2$ ')  # Ajout de la légende des axes et d'un titre
plt.show()  # Affichage de la figure 
```

#### Q5. Faire une figure représentant la fonction $sin(x)$ et la fonction $sin(x+\frac{\pi}{2})$ sur l'intervalle $[0, 2\pi]$. Chaque courbe doit être d'une couleur différente et le graphique doit être légendé.

```python
from math import pi
x = np.linspace(0, 2*pi, 100) # Liste des valeurs sur l'axe des abscisses
fx = np.sin(x) # Liste des valeurs sur l'axe des ordonnées
gx = np.sin(x+(pi/2))

fig, ax = plt.subplots() # Création de la figure (fig) et des axes (ax) de la figure
ax.plot(x, fx, label="$\\sin(x)$")           # Création de la courbe sur les axes de la figure
ax.plot(x, gx, label="$\\sin(x+\\frac{\pi}{2})$")

ax.set(xlabel='x', ylabel='y', title='Courbes sin')  # Ajout de la légende des axes et d'un titre
plt.legend()
plt.show()  # Affichage de la figure 
```

#### Q1. Visualisez ces 3 populations sur la période d'étude en traçant les 3 courbes sur le même graphique à l'aide de Matplotlib. N'oubliez pas le titre et les légendes.

```python
fig, ax = plt.subplots() 
ax.plot(annees, lievres, label="lièvres")           
ax.plot(annees, lynx, label="lynx")
ax.plot(annees, carottes, label="carottes")


ax.set(xlabel='Années', ylabel='Population', title='Evolution de la population')  # Ajout de la légende des axes et d'un titre
plt.legend()
plt.show()  # Affichage de la figure 
```
